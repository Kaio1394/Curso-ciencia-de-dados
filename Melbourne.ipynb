```python
# Trocando valores NaN por zero
data['Price'] = data['Price'].fillna(0)

# Preços com valores zerados
data.loc[data['Price'] == 0.0]
```

```python
# Casa com valor de preço maior
data.sort_values('Price', ascending=False)
```

```python
# Filtrando dados
filter = data.loc[(data['Rooms'] == 3) & (data['Type'] == 'h') & data['Price'] <= 200000]
filter
```

**Usando regex para filtrar os dados**

```python
# Filtrando com o método contains()
# 'turner st|turner Rd' -> contains('Turner St' or 'Turner Rd')
# 'flags'=re.I -> ignora letras maísculas
filter = data.loc[data['Address'].str.contains('turner st|turner Rd', flags=re.I)]
filter
```

```python
# '^59' -> Inicie como 59
filter = data.loc[data['Address'].str.contains('^59') & (data['Price'] <= 1000000)]
filter
```

```python
# Trocando valores da tabela
data.loc[data['SellerG'] == 'Nelson', 'SellerG'] = 'Kaio'
```

```python
# Confirmando valore trocado
data.loc[data['SellerG'] == 'Kaio']
```

```python
data.loc[data['SellerG'] == 'Kaio', 'Method'] = 'P'
data.loc[data['SellerG'] == 'Kaio']
```
